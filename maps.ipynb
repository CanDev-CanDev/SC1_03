```python
import folium
prov_geo = 'lpr_000b16a_e.shp'
folium_map = folium.Map(location=[55.284501, -98.1950807],
                    zoom_start=4,)
```

```python
folium_map
```

```python
import json
tweets = json.load(open("datasets/sample_tweets.json", 'r'))['statuses']
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="candev")
```

```python
for tweet in tweets:
    loc = tweet['user']['location']
    location = geolocator.geocode(loc)
    try:
        print(location.latitude, location.longitude)
        marker = folium.CircleMarker(location=[location.latitude, location.longitude])
        marker.add_to(folium_map)
    except:
        pass
```
